# Visualization of Loaded dataset

- To check the most used word in Ham sms and Spam SMS
- To visualize the percentage of Ham and Spam SMS

```python
# Stop words
stopwords = STOPWORDS
stopwords = list(stopwords)
STOPWORDS = nltk.corpus.stopwords.words('english')
stopwords = stopwords + STOPWORDS

ham_data = df[df.Label == 'ham']
spam_data = df[df.Label == 'spam']

ham_words = ' '
spam_words = ' '

for words in ham_data.Message:
    txt = words.lower()
    tokens = nltk.word_tokenize(txt)
    
    for word in tokens:
        ham_words =  ham_words + word + " "
for words in spam_data.Message:
    txt = words.lower()
    tokens = nltk.word_tokenize(txt)
    
    for word in tokens:
        spam_words = spam_words + word + " "

def gen_wordcloud(wordcloud):
    plt.figure(figsize=(10, 8))
    plt.imshow(wordcloud)
    plt.tight_layout(pad=0)
    plt.axis('off')
    plt.show()
    
print('\n')
print('\t\t\t\t HAM WORDS')
wordcloud = WordCloud(background_color='white', width=500, height=500, stopwords=stopwords,
                     max_words=500, max_font_size=50, random_state=42).generate(ham_words)

gen_wordcloud(wordcloud)

print('\t\t\t\t SPAM WORDS')
wordcloud = WordCloud(background_color='white', width=500, height=500, stopwords=stopwords,
                     max_words=500, max_font_size=50, random_state=42).generate(spam_words)
gen_wordcloud(wordcloud)
```

# Splitting the test and train data

```python
x_train, x_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=42)

print(x_train.shape); print(x_test.shape); print(y_train.shape); print(y_test.shape)

print('Referes that: Training set consists of {} samples of {} features'.format(x_train.shape[0], x_train.shape[1]))
print('Referes that: Test set consists of {} samples of {} features'.format(x_test.shape[0], x_test.shape[1]))
```

## Extracting N-grams from the Text data

```python
countVector = CountVectorizer(ngram_range=(2, 2), )
x_counts = countVector.fit(x_train.Message)

# Prepare training set
x_train_df = countVector.transform(x_train.Message)

# Prepare test set
x_test_df = countVector.transform(x_test.Message)
```

# Fit into Models
 
- ### Naive Bayes 
- ### KNN 


# 1. Naive Bayes Classifier

```python
naive_model = MultinomialNB()
naive_model.fit(x_train_df, x_train.Label)
predicted = naive_model.predict(x_test_df)
predictions = dict()

print(predicted)
accuracy = accuracy_score(x_test.Label, predicted)
predictions['Naive Bayes'] = accuracy * 100

con_mat = confusion_matrix(x_test.Label, predicted)
print("\n")
print("Accuracy of Naive Bayes Classifier is: {:.2f}%".format(accuracy * 100))
print('\n')
print('Confusion Matrix\n')
print(con_mat)

skplt.metrics.plot_confusion_matrix(x_test.Label, predicted, normalize=True)

plt.show()
```

# 2. K-Nearest Neighbor Classifier

```python
#KNN = KNeighborsClassifier(metric = 'euclidean')

knn_model = KNeighborsClassifier()

knn_model.fit(x_train_df, x_train.Label)
predicted = knn_model.predict(x_test_df)

print(predicted)
accuracy = accuracy_score(x_test.Label, predicted)

predictions['K-Nearest Neighbors algorithm'] = accuracy * 100
print("\nThe accuracy of K-Nearest Neighbors algorithm is: {:.2f}%".format(accuracy * 100))
conf_matrix = confusion_matrix(x_test.Label, predicted)

print('\n')
print('Confusion Matrix')
print("\n", conf_matrix)
skplt.metrics.plot_confusion_matrix(x_test.Label, predicted, normalize = True)
plt.show()
```

# Compare the Two Algorithms
